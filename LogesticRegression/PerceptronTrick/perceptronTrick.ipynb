```python
from sklearn.datasets import make_classification
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,
                        n_classes=2, n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
class Perceptron:
    def __init__(self,X,y):
        self.X=X
        self.y=y
        self.learningRate=0.1
    def step(self,x):
        return 1 if x>0 else 0
    def fit(self):
        self.X=np.insert(self.X,0,1,axis=1)
        weights=np.ones(self.X.shape[1])
        for i in range(1000):
            j=np.random.randint(0,100)
            y_hat=self.step(np.dot(self.X[j],weights))
            weights=weights+self.learningRate*(self.y[j]-y_hat)*self.X[j]
        return weights[0],weights[1:]
```

```python
model=Perceptron(X,y)
```

```python
intercept_,coef_=model.fit()
```

```python
intercept_
```

```python
coef_
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
class SigmoidPerceptron:
    def __init__(self,X,y):
        self.X=X
        self.y=y
        self.learningRate=0.1
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    def fit(self):
        self.X=np.insert(self.X,0,1,axis=1)
        weights=np.ones(self.X.shape[1])
        for i in range(1000):
            j=np.random.randint(0,100)
            y_hat=self.sigmoid(np.dot(self.X[j],weights))
            weights=weights+self.learningRate*(self.y[j]-y_hat)*self.X[j]
        return weights[0],weights[1:]
```

```python
model=SigmoidPerceptron(X,y)
```

```python
intercept_,coef_=model.fit()
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
